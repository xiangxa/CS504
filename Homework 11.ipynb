```python
#1- Download and import the BMI_Regional_Data.xls file posted in the lesson 11 folder;
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
data1 = pd.read_excel('BMI_Regional_Data.xlsx')
data1.head()
```

```python
#2- Create a dataset which slices the first 100 rows from the original imported dataset;
data100 = data1.head(100)
print(data100)
```

```python
#3- Determine how many missing values are there? How many missing values per column/variable;
data100.isnull().sum().sum()
#There is 122 Total missing value in the 100 rows
data100.isnull().sum()
#Out of the 122 missing value, 100 come from ISO, 2 come sex, 5 come from Year, 6 come from Mean BMI
#6 come from Mean BMI lower 95% uncertainty interval, 1 comes from Mean BMI upper 95% uncertainty interval                                       1
#and 1 comes from Prevalence of BMI>=30 kg/m² (obesity) and 1 comes from Prevalence of BMI>=35 kg/m² (severe obesity) 
```

```python
#5- From the sliced dataset containing the first 100 rows, impute the missing values for the 'Year' variable,
#by replacing them with their correct values using specific cell locations, or an algorithm;¶


data100["RecordID"] = data100.index

for index,row in data100.iterrows():
    if data100["Country/Region/World"]=="Central and Eastern Europe":
        data100["RecordID"]= data100["RecordID"]
    elif data100["Country/Region/World"]=="Central Asia, Middle East and North Africa":
        data100["RecordID"]= data100["RecordID"]-41
    else:    
        data100["RecordID"] = data100["RecordID"]-84
        
data100["Year"].fillna((2016-data100.RecordID), inplace=True)
data100.head(100)
```
