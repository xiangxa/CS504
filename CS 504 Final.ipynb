```python
#Exercise 3:
#Develop and code a guessing program, which asks the user to generate a random number in the range (1:100). 
#The program is not being informed of the value of the randomly generated number so that it can guess the number. 
#Once you ahve generated the number develop your guessing algorithm as follows: 
#the program will begin guessing by selecting a number in the middle of the range. 
#The program will present this number to the user, the user will then indicate whether the number is too high or too low. 
#If it is too high then the program will present a number that is one integer lower, and if it too low then the program will present a number that is 1 integer higher. 
#For example if the program guesses 54 and the user indicates too high then the next number the program will present is 53
import numpy as np
gen=round(np.random.rand()*100)
print ("The random number you generate is "+ str(gen))

guess = round(np.random.rand()*100)
print ("my guess is "+ str(guess) + ".")
while gen != guess:
    indicate = str(input("Is it too high or too low or I am right: "))
    if "high" in indicate:
        print ("my next guess is :"+ str(guess-1))
        guess = guess-1
    elif "low" in indicate:
        print ("my next guess is :"+ str(guess+1))
        guess = guess+1
    else: print ("I got your number")
```
